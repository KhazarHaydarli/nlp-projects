```python
text_train = train_data[0].replace("movie0.flac", "")
text_train = text_train.replace("\n", "")
text_train
# train_data[0] = train_data[0].map(lambda x: x.rstrip('movie0.flac'))
```

```python
texts_train = []
for i in range(0, len(train_data)):
  txt = train_data[i].replace(f'movie{i}.flac', '')
  txt = txt.replace("\n", "")
  texts_train.append(txt)
```

```python
audio_train = []
for i in range(0, len(train_data)):
  ses = train_data[i].split(" ")[0]
  audio_train.append(ses)
```

```python
a = {}
for i in range(len(audio_train)):
  a[audio_train[i]] = texts_train[i]
```

```python
df_train = pd.DataFrame(a.items())
df_train
```

```python
df_train.rename(columns = {0:'audio', 1:'text'}, inplace = True)
df_train
```
